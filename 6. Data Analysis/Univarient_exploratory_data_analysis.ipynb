# 1. Categorical Data

## Countplot

```python
import seaborn as sns
```

### Monitoring the survival of people

```python
sns.countplot(x = 'Survived', data=df)
df['Survived'].value_counts()
```

Alternative way of plotting the same graph.
* find the value count for each category and plot that on bar kind of graph

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

### Monitoring the devision of people in terms of class

```python
sns.countplot(x='Pclass', data = df)
```

```python
df['Pclass'].value_counts()
```

Alternative way of plotting the same graph.
* find the value count for each category and plot that on bar kind of graph

```python
df['Pclass'].value_counts().plot(kind='bar')
```

### Monitoring where exactly people has embarked from

# Numerical Data

## Histogram

* We can make numerical data behave like a categorical data using histogram by creating bins.
* let's say the minimum age on board of the passenger is 0 and the maximum is 80 we can create 8 bins that will range from 0-10, 10,-20 and so on.

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=15)
```

## Distplot

This gives us PDF i.e probability density function

We can get the skewness with this plot

```python
sns.distplot(df['Age'],bins=30)
```

## Boxplot

this gives us 5 number summary:
* Median (50 percentile)
* Percentile value (*Q1 = 25th percentile value and *Q3 = 75th percentile value)
* Inter Quartile Range(IQR) are values in between Q1 and Q3
* *Minimum is calculated by (Q1-1.5*IQR) 
* *Maximum is calculated by (Q3+1.5*IQR)


```python
sns.boxplot(x='Age', data=df)
```
