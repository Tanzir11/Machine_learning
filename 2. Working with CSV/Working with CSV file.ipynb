```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/JangirSumit/data_science/master/15th%20June%20Assignments/case%20study%201/BX-Books.csv"
headers = {"user-Agent": "Mozilla/5.0 (Macintosh; Intel Max OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

data_frame = pd.read_csv(data)


# Save the DataFrame to a CSV file
data_frame.to_csv('BX-Book.csv')  # Optional: exclude the index
```

### Sep Parameters

The default value of sep = "," which means when pandas is reading the csv file then it is seprating everything by the "," which works for csv file because it's a "comman" separated value but when we are talking about "tsv" which basically "tab" separated vaule we will have the modify sep parameter from "," to "\t"

```python
pd.read_csv("reference.tsv", sep='\t')
```

### Converters

Let's say we want to convert Royal Challengers name to RCB and so on, so we will do it as follows

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
```

```python
import pandas as pd

pd.read_csv('IPL_Matches_2008_2022.csv', converters={'Team1':rename})
# so basically converters={'Team1':rename} is going inside the Team1 column and applying the logic for function rename
```

### na_values parameter

We can easisy deal with prebaked nan values that is not an issue. the issue arises when we get "-", "na" and etc instead of NaN. To deal with this we can decide what are the parameters we consider Nan and that's what we are going to see as of now.

```python
pd.read_csv('demo_dataframe.csv')
```

```python
pd.read_csv('demo_dataframe.csv', na_values=['John Smith','12345'])
```

So now John Smith is being converted to NaN value and 12345 as well.

### Loading a huge dataset in chunks

```python
pd.read_csv('demo_dataframe.csv')
```

```python
dfs = pd.read_csv('demo_dataframe.csv', chunksize=2)
```

now anything i want to do i would do inside a loop, as follows.

```python
for chunks in dfs:
    print(chunks.shape)
```

So basically our dataset is now divided in a chunk of (2,4) i.e 2 rows and 4 columns.
