### Sigmoid function:
**Sigmoid Function**

The sigmoid function is a mathematical function with a characteristic "S"-shaped curve. It's widely used in machine learning and neural networks. Here's the equation:

$$S(x) = \frac{1}{1 + e^{-x}}$$

**Explanation:**

* **x:** The input value 
* **e:** Euler's number (approximately 2.71828)
* **S(x):** The output of the sigmoid function (a value between 0 and 1) 


```python
import matplotlib.pyplot as plt
import numpy as np

#Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate input values
x = np.linspace(-10, 10)

# Calculate sigmoid output
y = sigmoid(x)

# Plot the graph
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Function")
plt.savefig("sigmoid_plot.png")
```

* Sigmoid function ranges between 0-1.
* it's a derivatible function is easier to calculate for gradient descent calculation
